# Question 3
Write a Python program to plot a scatter plot using discrete values to generate balls of
different sizes.


```python
import numpy as np
import matplotlib.pyplot as plt

# generate random data
x = np.random.randint(1, 11, size=40)
y = np.random.randint(1, 11, size=40)
sizes = np.random.randint(20, 200, size=40)

# create colormap for sizes
cmap = plt.cm.get_cmap('RdYlBu')
norm = plt.Normalize(vmin=min(sizes), vmax=max(sizes))

# create scatter plot
fig, ax = plt.subplots()
scatter = ax.scatter(x, y, s=sizes, c=sizes, cmap=cmap, norm=norm)

# set title and axis labels
ax.set_title('Scatter Plot with Different-Sized and Different-Colored Circles')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# add colorbar
cbar = plt.colorbar(scatter, ax=ax)
cbar.ax.set_ylabel('Sizes')

# show the plot
plt.show()
```

# Question 6
Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.

Sample Financial data (fdata.csv):

Date,Open, High, Low, Close



10-03-16,774.25,776.065002,769.5,772.559998

10-04-16,776.030029,778.710022,772.890015,776.429993

10-05-16,779.309998,782.070007,775.650024,776.469971

10-06-16,779,780.47998,775.539978,776.859985

10-07-16,779.659973,779.659973,770.75,775.080017

```python
import matplotlib.pyplot as plt

dates = ['10-03-16', '10-04-16', '10-05-16', '10-06-16', '10-07-16']
opens = [774.25, 776.030029, 779.309998, 779, 779.659973]
highs = [776.065002, 778.710022, 782.070007, 780.47998, 779.659973]
lows = [769.5, 772.890015, 775.650024, 775.539978, 770.75]
closes = [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]

plt.plot(dates, opens, color='red', label='Open')
plt.plot(dates, highs, color='blue', label='High')
plt.plot(dates, lows, color='green', label='Low')
plt.plot(dates, closes, color='orange', label='Close')

plt.title('Financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016')
plt.legend()

plt.xticks(rotation=45)  # rotate x-axis labels for better readability

plt.tight_layout()
plt.show()
```
